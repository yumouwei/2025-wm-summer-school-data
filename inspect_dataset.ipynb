# Inspect dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cmod_dataset = pd.read_csv('./data/cmod_dataset.csv')
```

```python
print(f"Total number of shots: {len(cmod_dataset['shot'].unique())}")
print(f"Number of disruptive shots: {len(cmod_dataset.loc[cmod_dataset['time_until_disrupt'].notna()]['shot'].unique())}")
print(f"Number of non disruptive shots: {len(cmod_dataset.loc[cmod_dataset['time_until_disrupt'].isna()]['shot'].unique())}")
```

```python
shot = cmod_dataset['shot'].unique()[1000]
shot_data = cmod_dataset.loc[cmod_dataset['shot']==shot]

fig, axes = plt.subplots(2, sharex=True)
axes[0].plot(shot_data['time'], shot_data['ip'], label='ip')
axes[0].plot(shot_data['time'], shot_data['ip_prog'], linestyle='--', label='ip_prog')
axes[0].legend()
axes[1].plot(shot_data['time'], shot_data['time_until_disrupt'])
axes[1].set_ylabel('time_until_disrupt (s)')
axes[1].set_xlabel('time (s)')
plt.suptitle(shot)
plt.show()
```
